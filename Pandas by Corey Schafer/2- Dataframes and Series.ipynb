## Dataframes
we might see defintion of dataframe, 2-dimensional datastructure  
but it is basically a table with rows and columns   
- columns are keys of dictionaries and   
- rows are values in the list of keys

```python
import pandas as pd
```

```python
#  dummy dictionary to convert it into dataframe

data = {
    'name': ['Alice', 'Bob', 'Charlie', 'David'],
    'age': [25, 32, 18, 47],
    'shape': ['New York', 'Paris', 'London', 'San Francisco']
}
```

```python
# converting into dataframe 

df= pd.DataFrame(data)
```

```python
df

# 0,1,2,3 are indexes, this a column added by the pandas
# we can create our own index column, will see later
# index column generated by the pandas is unique but
# we can set any column as index column if it has unique values of not
```

## Series

we might see definition online, like 1-dimensional data structure

it is just a list with additional functionalities of pandas  
#### We can think it, a dataframe contains multiple object of series

```python
df["name"]
```

#### Dot notation to get the series out of the dataframe

Dot notaion and bracket notation gives the same results

bracket notation is preferred because, there is a chance that our column can be named as the **pandas dataframe's** attribute or function e.g. **shape**, shape can be our column and shape is also the attribute of the **dataframe** 

- if we have a column name same as the attribute or function, Pandas will give preference to the attribute or function, So, bracket notation is more useful

- We cannot apply dot notataion to the column names that have spaces

```python
df.name
```

```python
# printing "shape" column which is also the attribute of the dataframe, 
# pandas gives results on the attribute not the column name
df.shape
```

## iloc and loc 

#### indexed-location and location

loc and iloc are indexing operators used to select rows and columns from a DataFrame. They are actually attributes of the DataFrame object and provide a way to select rows and columns by label or by integer position, respectively.

```python
# let's first discuss about iloc

# iloc is an integer-based indexing method that selects rows and columns based on their integer position in the DataFrame.


# we can pass a integer in iloc to get result of row
df.iloc[1]

# here's the row 1 
# look closely, in this scenario index column is the names of the columns and 2nd column contains values of the rows
```

```python
# checking type of returned result of iloc
type(df.iloc[1])

# it is giving us the series object
```

```python
# we can also pass a list to iloc to print multiple rows 
df.iloc[[0,1]]

# it is giving us all column with 2 rows 0 and 1
```

```python
# as we know iloc is used for integer based indexing, we can pass specify the columns as well just like the rows 
df.iloc[[0,1],[0,1]]

# this will give us 0 and 1 rows and 0 and 1 columns which are name and age
```

```python
# if we want to get only a single value with corresponding column we can do this
df.iloc[0,2]

# it is giving me the value of 0th row and 3rd column which is shape/city in our df
```

#### loc

```python
# just like the iloc operator we can specify the rows and columns in the loc operator as well 
df.loc[0]

# it gives us the same result as iloc
```

```python
# Remeber loc is label based operator, we can pass it a list of integers to give results of only those result and 
# we can also pass it columns 
df.loc[[0,1]]
```

```python
df.loc[[1,3]]
```

```python
df.loc[[1,3],["name","age"]]
```

```python
# if we want to get only a single value with corresponding column we can do this
df.loc[0,"name"]

# it is giving me the value of 0th row and name which is shape/city in our df
```

## Slicing for loc and iloc

slicing is pretty much same like "python" but the only difference is, end point of pandas slicing is inclusive only in case of **loc** (unlike python)  and in case of **iloc** it is not inclusive rather exclusive (like python)

```python
df2.loc[0:4, "Hobbyist": "Country"]
```

```python
df2.iloc[0:4, 0:6]
```
