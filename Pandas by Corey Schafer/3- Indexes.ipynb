## Setting up custom index column
Most of the cases, pandas does not restricts to have a custom index column which have duplicated values but but it might be the case where pandas restricts to set custom index column without any duplicate value.  
to set the index we use **set_index()** function  
resetting index: **reset_index()**

```python
# setting "Respondent" column as indexed column 
df.set_index("Respondent")

# as you can see, "Respondent" column is now below from all other columns in alignment, it is set as index column 
# let's discuss the benefits of setting a column as index
```

```python
# if we print our original dataset, we can see pandas did not change the dataset,
df

# still index column in present, but if we want to set the index right on the main dataset, we set "inplace" argument as true
```

```python
df.set_index("Country", inplace=True)
```

```python
df

# Now, if I print the dataframe, "Country" column in set as index on the original dataset
```

```python
# we can print the index column just by using the "index" attribute of pandas

df.index
```

```python
df.reset_index()
```

```python
df.index
```

```python
df.reset_index(inplace=True)
```

```python
df.index
```

#### Setting index
If the index is set, we cannot get the rows by passing any integer in loc operator to get that particular row,  
rather we need to pass the values from the column which is set as index. 

```python
# let's the set the df dataframe index as "Country"
df.set_index("Country", inplace = True)
```

```python
df
```

```python
# Now we cannot search of row numbers in loc by passing integer or list of integers in square brackets
df.loc[0]

# because we have removed the default index column of pandas, and above syntax works on the default indexed column
```

```python
# we can get value by passing country name 
df.loc["Pakistan"]
```

```python
# OR we can set the index to the "Respondent" column and than we can search any row by passing the integers into loc, 
# because respondent column have unique ids already So, lets try it

df.reset_index(inplace=True)

df.set_index("Respondent",inplace=True)

print(df.loc[1]) 
# the result is in the Series Type 

print(type(df.loc[1]))
```

```python
# we can get multiple rows throug index as well like: 
df.loc[[1,2,3]]
```

```python
# We can also get specified columns like
df.loc[1,"MainBranch"]
```

```python
# OR we can get multiple columns as well 
df.loc[1,["MainBranch","Hobbyist"]]
```

```python
# OR we can get multiple rows as well and columns as well
df.loc[[1,2], ["MainBranch","Hobbyist"]]
```
