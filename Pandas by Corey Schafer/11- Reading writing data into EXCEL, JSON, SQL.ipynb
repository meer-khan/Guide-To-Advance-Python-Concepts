## Tab seperated format
using the "to_csv()" function of pandas dataframe, we can create any type of seperated files, like tab seperated file or maybe "-" dash seperated or "_" underscore sepeared file using the **sep** argument

and if we want to read tab sepeated format, we will read with "read_csv()" function and add the "sep" argument and pass the "\t" to read it into the tab seperated format

```python
pak_df.to_csv("CSVs\Pakistan_survey_results.tsv", sep="\t")
```

## JSON format

```python
pak_df.to_json("CSVs\Pakistan_survey_results.json")
```

```python
# making json into  single dicntionary like not the bunch of implicit dictionaries like
# orient argument make it seperate dictionary like like 
# lines argument will display each list in sepearte line
pak_df.to_json("CSVs\Pakistan_survey_results_list_like.json", orient='records', lines=True)
```

```python
# Reading json file
# Reading the unknown json file, we need to try different arguments to fit that json format and read it through pandas
# but for our file we know the format so we will pass extact those parameters to read it

pak_df_json = pd.read_json("CSVs\Pakistan_survey_results_list_like.json", orient='records', lines=True)
```

## Reading SQL data

```python
from sqlalchemy import create_engine
import psycopg2
```

```python
engine = create_engine("postgresql://dbuser:dbpass@localhost:5432/sample_db")
```

```python
# {'fail', 'replace', 'append'}, these 3 values can be passed to if_exists, check documentation of this function
pak_df.to_sql("sample_table", engine, if_exists='replace')
```

```python
# Reading from sql data
pak_df = pd.read_sql("sample_table", engine, index_col="Respondent")
```

```python
# Applying query on the database 
pd.read_sql_query("SELECT * FROM sample_table", engine , index_col="Respondent")
```
