## Filtering

Filtering is all about getting the specific data from columns and apply multiple gate operations on it like OR, AND, NOT

```python
# Let's filter the records, where "Country" column has value "Pakistan"
df["Country"] == "Pakistan"

# This is giving us all the rows (Series) and shows results in boolean form, either "Country" column has value "Pakistan" or not
# We can convert them into actual dataframes
```

```python
results = df["Country"] == "Pakistan"
df[results]

# this is giving us all those rows that has value "Pakistan" in "Country" column
```

```python
# Short hand notation of filtering: 

df[df["Country"] == "Pakistan"]
```

## Confusing Pandas

pandas is very confusing, because we can pass multiple things to multiple objects and they produce different results,Sometimes we need to use double pair of brackets and sometimes we use single pair of brackets etc, let's see an example right below

```python
# Passing the boolean results of Country as Pakistan into the loc operator, gives us the acutal dataframe
df.loc[results]

# Really confusing right? but thats how pandas work
```

```python
# We can do another thing, we can use ~ tilde operator in loc operator to print dataframe where Country is not Pakistan,
# ~ tilde act as a NOT operator, that converts True into False and vice versa
df.loc[~results]

# As you can see down there in the results, All the rows where Country is not Pakistan is given as resultant
```

```python
# We can get the spcified column as well

df.loc[results, "Hobbyist"]
```

```python
# We can get multiple columns as well

df.loc[results, ["Hobbyist","Employment"]]
```

## AND & OR | operators

```python
# To use the AND or OR operators we need to enclose our conditions into () round pair of brackets, 
# otherwise Pandas throws error

results = (df["Country"] == "Pakistan") & (df["Gender"] == "Man")
```

```python
df.loc[results]
```

```python
df.loc[results, ["Country", "Gender", "LanguageWorkedWith", "CompTotal"]]
```

## isin() function

```python
# isin() method is used to apply the filter on multiple values rather than us OR | operator excessively

countries = ["Pakistan", "India"]
results = (df["Country"].isin(countries)) & (df["Gender"] == "Man")
```

```python
df.loc[results, ["Country", "Gender", "LanguageWorkedWith", "CompTotal"]]

# Here we have results of Countries, Pakistan and India where Gender is Man
```

## string methods

```python
# Let's search for the languages "Python" on which stackoverflow users worked on
# if we see the dataset, "LanguageWorkedWith" column contains multiple languages seperated by the semicolon ; 
# we cannot directly filter our like LanguageWorkedWith == "Python"
# Either we need to apply some regex over it to find people who use "Python" or Pandas provides us 
# .str.contains() function to search into the string 

# we gave parameter "na" and set it as False because, dataset has NaN values, and on NaN values, Pandas will give error, so 
# we set it False

filt = (df["Country"] == "Pakistan") & (df["LanguageWorkedWith"].str.contains("Python", na=False))
```

```python
re = df.loc[filt,  ["Country", "Gender", "LanguageWorkedWith", "CompTotal", "YearsCodePro"]]
```
